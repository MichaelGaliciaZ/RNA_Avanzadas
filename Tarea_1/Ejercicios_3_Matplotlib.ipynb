## Ejercicio 3

Crear un objeto  y agregar dos ejes (axis) , en las siguientes locaciones [0,0,1,1] y [0.2,0.5,0.2,0.2]  


```python
fig=plt.figure(figsize=(8,8))
ax1=fig.add_axes([0,0,1,1])
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.set_title("Title")

ax2=fig.add_axes([0.2,0.5,0.2,0.2])
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
```

- Graficar x,y en ambos ejes y llamar a la figura  

```python
ax1.plot(np.linspace(0,100,100),np.linspace(0,200,100))
ax2.plot(np.linspace(0,100,100),np.linspace(0,200,100))
fig
```

## Ejercicio 4

Crear un objeto  y agregar dos ejes (axis) , en las siguientes locaciones [0,0,1,1] y [0.2,0.5,.4,.4]  


```python
fig2=plt.figure()
ax1=fig2.add_axes([0,0,1,1])
ax2=fig2.add_axes([0.2,0.5,.4,.4])
```

Graficar x,y en el eje de menor tamaño  y x,z en eje principal, poner particular atención en los límites de las figuras 

```python
ax1.plot(np.linspace(0,100),np.linspace(0,100)**2)
ax2.plot([20,22],[40,45])
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
ax2.set_title("zoom")
ax1.set_xlabel("X")
ax1.set_ylabel("Z")
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
fig2
```
